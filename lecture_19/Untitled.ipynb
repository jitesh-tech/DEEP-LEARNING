```python
# NOW WE ARE GOING TO LEARN ABOUT MINIBATCH GRADIENT DESCENT
# Mini-Batch Gradient Descent is an optimization algorithm where the model parameters (weights and biases) are updated after
# computing the gradient on a small subset (mini-batch) of the training data,
# instead of the whole dataset (Batch GD) or just one sample (SGD).
```

```python
# ✅ Why Mini-Batch is Better (Advantages)

# 1️⃣ Faster than Batch GD
# Updates happen more frequently.

# 2️⃣ More Stable than SGD
# Averaging over a batch reduces noise.

# 3️⃣ Efficient for Large Datasets
# Works well with GPUs (parallel computation).

# 4️⃣ Balanced Convergence
# Converges faster and more smoothly.

# 5️⃣ Standard Method in Deep Learning
# Almost all modern deep learning frameworks use mini-batches.
```

```python
# 🎯 Why It Is Preferred in Deep Learning

# Deep neural networks have:

# Millions of parameters

# Large datasets

# Mini-Batch Gradient Descent provides:
# ✔ Efficient computation
# ✔ Stable learning
# ✔ Faster convergence
# ✔ Better generalization
```
