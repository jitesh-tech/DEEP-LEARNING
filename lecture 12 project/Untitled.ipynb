```python
import tensorflow 
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Flatten
```

```python

(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
```

```python
x_train.shape
```

```python
# today we are going to train our deep learning ann model on mnist data set to identify the no
```

```python
print(x_train)
```

```python
x_test.shape
```

```python
# now we are going to convert that pixel value in image
import matplotlib.pyplot as plt
plt.imshow(x_train[2])
```

```python
# reason behind to make all the points or values close to each other becasuse whenever the values are closer are each other the
# then we weight can be easily find and convergence process is became faster
```

```python
x_train = x_train/255
x_test = x_test/255
```

<!-- now we are going to make our deep learning ann model -->
<!--  -->

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten

model = Sequential()

# Convert 28x28 image into 1D (784)
model.add(Flatten(input_shape=(28, 28)))

#  first Hidden layer
model.add(Dense(128, activation='relu'))

# Output layer having 10 nodes with activation fun is softmax 
model.add(Dense(10, activation='softmax'))
```

```python
model.summary()
```

```python
# here in fist hidden layer : 100480 trainable parametres and in output layer 1290 trainable parametres 
# 784: show no of inputs in inputs layers  128: shows no of nodes in hidden layers and 10 show no of nodes in outputs layer
```

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
model.fit(x_train, y_train, epochs=10, validation_split=0.2)
```

```python
y_prob = model.predict(x_test)
```

```python
print(y_prob)
```

```python
y_pred = y_prob.argmax(axis=1)
```

```python
from sklearn.metrics import accuracy_score
model_acc = accuracy_score(y_test, y_pred)
```

```python
# so finally our model prediction accurracy sccore is : 97 percent which mean our prediction is 97 percent is match with y_test
# data which we kept for match our resutlt so that our model is prediction well or not 
```
